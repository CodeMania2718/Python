```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('data/filtered_cpu_5_ips_latest.csv',parse_dates=['ingestion_time'])
```

```python
df.columns
```

```python
df.drop(['Unnamed: 0', 'id',],axis=1,inplace=True)
df.sort_values(by='ingestion_time',inplace=True)
df.set_index('ingestion_time',inplace=True,drop=False)
```

```python
print(df['ip'].nunique())
df['ip'].unique()
```

```python
start_time='2023-05-09 00:04:35'
end_time='2023-05-09 01:04:35'
df1=df.loc[(df.index >= start_time) & (df.index < end_time)]
df1[df1['ip']=='10.80.80.36']['value'].plot(grid=True,figsize=(30,9),marker="o")
```

```python
df1.groupby('ip')['value'].plot(grid=True,figsize=(30,9))
plt.legend(loc='upper right')
plt.show()
```

```python
gd=df1.groupby('ip').agg({"value":['max', 'mean']})
gd.columns=[f"{i}_{j}" for i,j in gd.columns]
gd
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
```

```python
scaler = StandardScaler()
X_norm=scaler.fit_transform(gd[['value_max','value_mean']])
kmeans = KMeans(random_state=42)
visualizer = KElbowVisualizer(kmeans, k=4, metric='distortion', timings=1)
visualizer.fit(X_norm)
```

```python
kmeans = KMeans(n_clusters=3,random_state=42)
kmeans.fit(X_norm)
gd['KMeans']=kmeans.labels_
gd
```
