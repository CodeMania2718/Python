```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('data/filtered_memory_5_ips.csv',parse_dates=['ingestion_time'])
```

```python
df.columns
```

```python
df.drop(['Unnamed: 0', 'id',],axis=1,inplace=True)
df.sort_values(by='ingestion_time',inplace=True)
df.set_index('ingestion_time',inplace=True,drop=False)
```
