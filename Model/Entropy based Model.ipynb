```python
from scipy.signal import periodogram

def spectral_entropy(x, freq, nfft=None):    
    _, psd = periodogram(x, freq, nfft = nfft)   
    # calculate shannon entropy of normalized psd
    psd_norm = psd / np.sum(psd)
    entropy = np.nansum(psd_norm * np.log2(psd_norm))
    return -(entropy / np.log2(psd_norm.size))
```

```python
import matplotlib.pyplot as plt
window = 24
nfft = None
df = pd.DataFrame(data=df2, columns=['average_percent_memory_used'])
df['x_roll_se'] = df['average_percent_memory_used'].rolling(window).apply(lambda x: spectral_entropy(x,freq=100,nfft=nfft))
plt.figure(figsize=(12,3))
plt.plot(df['x_roll_se'])
plt.ylabel('Rolling Spectral Entropy')
plt.title(f'window size:{window}, FFT size:{nfft}')    
plt.show()

x = df['x_roll_se'].max()
y = df['x_roll_se'].min()
mean_val = df['x_roll_se'].mean()
std_val = df['x_roll_se'].std()
print(x)
print(y)
print(mean_val)
print(std_val)

plt.figure(figsize=(12,3))
plt.plot(df2['average_percent_memory_used'])
```
